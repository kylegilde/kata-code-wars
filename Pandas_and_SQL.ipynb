Desired Output

The resulting set should look similar to the following:

```
date       | count | total
-----------+-------+-------
2017-01-26 |    20 |    20
2017-01-27 |    17 |    37
2017-01-28 |     7 |    44
2017-01-29 |     8 |    52
```

date - (DATE) date

count - (INT) number of posts for a date

total - (INT) a running (cumulative) number of posts up until a date

````python
from io import StringIO
import pandas as pd
df = pd.read_csv(StringIO(s), sep="\t")
df.info()
(
    df
    .assign(date=lambda df: df.created_at.astype("datetime64").dt.date)
    .groupby("date")
    .size()
    .rename("n")
    .reset_index()
    .assign(total=lambda df: df.n.cumsum())
)
````

```
SELECT
    date,
    count,
    CAST(SUM(count) OVER(ORDER BY date ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS INTEGER) AS total
FROM (
    SELECT
      CAST(created_at AS DATE) AS date,
      COUNT(*) AS count
    FROM
      posts
    GROUP BY 1
) AS p
```
